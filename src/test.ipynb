```python
from fin import Stock
```

```python
def report(ticker: str):
    today = Stock.today(ticker)
    one_month = Stock.months(ticker, num_months=60)
    ytd = Stock.ytd(ticker)
    return ytd.chart | one_month.chart | today.chart
```

```python
report('AAPL')
```

```python
report('NVDA')
```
